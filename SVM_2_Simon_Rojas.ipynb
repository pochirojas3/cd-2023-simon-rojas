```python
import numpy as np
import pandas as pd
import seaborn as sb
import os
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/adiacla/bigdata/master/titanic.csv")
df
```

```python
#Verificar nulos
df.isna().sum()
```

```python
df = df.drop('Cabin', axis = 1)
df
```

```python
df = df.dropna()
df
```

```python
df.info()
```

```python
freq = df.groupby(['Sex']).count()
freq
```

```python
freq = df.groupby(['Embarked']).count()
freq
```

```python
df["Sex2"] = 0
df.Sex2.loc[df.Sex=="female"] = 1
df.Sex2.loc[df.Sex=="male"] = 0
df
```

```python
df["Embarked2"] = 0
df.Embarked2.loc[df.Embarked=="S"] = 2
df.Embarked2.loc[df.Embarked=="Q"] = 1
df.Embarked2.loc[df.Embarked=="C"] = 0
df
```

```python
#Selección de variables
predictores = df[['Pclass','Age','SibSp','Sex2','Embarked2']]
objetivo = df.Survived
```

```python
pred_train, pred_test, res_train, res_test = train_test_split(predictores, objetivo, test_size=0.2)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, recall_score, precision_score

SVM_Model = SVC(gamma='auto', kernel='rbf')
SVM_Model.fit(pred_train, res_train)
```

```python
print (f'Precisión: {SVM_Model.score(pred_test, res_test):.4f}')
```

```python
predicciones = SVM_Model.predict(pred_test)
precision = precision_score(res_test, predicciones)
exhaustividad = recall_score(res_test, predicciones)
exactitud = accuracy_score(res_test, predicciones)

# Resultados
print('Exactitud:', exactitud)
print('Precisión:', precision)
print('Exhaustividad:', exhaustividad)
```
